```python
def make_prompt(text):
    return (
        "You are given an essay. Analyze it carefully and determine whether each of the following Big Five personality traits is present or not. "
        "For each trait, output 1 if the trait is clearly present in the essay, and 0 if it is not. Output only binary values (0 or 1). "
        "Format your response strictly as:\n\n"
        "O: <0 or 1>  # Openness to Experience: Adventurous, enjoys novel experiences, creative, curious, interested in arts, imaginative, open to new ideas, thinks abstractly, embraces diversity\n"
        "C: <0 or 1>  # Conscientiousness: Organized, responsible, goal-oriented, disciplined, reliable, pays attention to detail\n"
        "E: <0 or 1>  # Extraversion: Outgoing, energetic, talkative, assertive, enjoys social interaction, shows warmth and enthusiasm\n"
        "A: <0 or 1>  # Agreeableness: Compassionate, cooperative, empathetic, puts others' needs first, enjoys helping and caring for others\n"
        "N: <0 or 1>  # Neuroticism: Experiences negative emotions like anxiety, moodiness, self-doubt, easily stressed, emotionally unstable, overreacts to minor issues\n\n"
        "Essay:\n"
        f"{text}"
    )
```

**GETTING RESPONSE OF LLAMA-3.3-70B-VERSATILE**

```python
import re
import time
import openai

results = []

for i, essay in enumerate(essays, 1):
    prompt = make_prompt(essay)

    retries = 0
    max_retries = 5
    while True:
        try:
            response = openai.ChatCompletion.create(
                model="llama-3.3-70b-versatile",
                messages=[{"role": "user", "content": prompt}],
                temperature=0.7
            )
            text = response['choices'][0]['message']['content']
            print(f"Response for essay {i}:\n{text}\n")
            results.append(text)
            break  # successful request, exit retry loop

        except openai.error.RateLimitError as e:
            wait = 2 ** retries
            print(f"Rate limit hit. Waiting {wait} seconds before retrying...")
            time.sleep(wait)
            retries += 1


    # Optional: Short pause to avoid hitting limits too fast
    time.sleep(1)
```

**EXTRACTION OF OCEAN SCORES FROM GENERATED OUTPUTS OF MODELS**

```python
 import re
import csv

# Flatten tuples if needed
res = [r[0] if isinstance(r, tuple) else r for r in results]

pattern = (
    r"O:\s*(\d+).*?"
    r"C:\s*(\d+).*?"
    r"E:\s*(\d+).*?"
    r"A:\s*(\d+).*?"
    r"N:\s*(\d+)"
)


with open('versatile.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['O', 'C', 'E', 'A', 'N'])
    for i, text in enumerate(res):
        match = re.search(pattern, text, re.DOTALL | re.IGNORECASE)
        if match:
            scores = list(map(int, match.groups()))
            writer.writerow(scores)
        else:
            print(f"No OCEAN scores found in entry {i + 1}")
            writer.writerow([0, 0, 0, 0, 0])
```

```python
import re
import csv
from datasets import load_dataset

# Load first 100 samples from Essay dataset
dataset = load_dataset("jingjietan/essays-big5", split="train[:100]")

# Extract fields
essays = dataset['text']
orig_scores = list(zip(
    dataset['O'],
    dataset['C'],
    dataset['E'],
    dataset['A'],
    dataset['N']
))

# Flatten predictions if needed
res = [r[0] if isinstance(r, tuple) else r for r in results3]
res = res[:len(orig_scores)]  # Match length

# Regex to extract OCEAN scores from model output
pattern = r"O:\s*(\d+).*?C:\s*(\d+).*?E:\s*(\d+).*?A:\s*(\d+).*?N:\s*(\d+)"

# Write to CSV
with open('meta40.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Original Sentence',
                     'Pred_O', 'Pred_C', 'Pred_E', 'Pred_A', 'Pred_N',
                     'True_O', 'True_C', 'True_E', 'True_A', 'True_N'])

    for i, text in enumerate(res):
        match = re.search(pattern, text, re.DOTALL | re.IGNORECASE)
        sentence = essays[i]
        true_scores = [round(float(orig_scores[i][j]) , 2) for j in range(5)]

        if match:
            pred_scores = list(map(int, match.groups()))
        else:
            print(f"No OCEAN scores found in entry {i + 1}")
            pred_scores = [0, 0, 0, 0, 0]

        writer.writerow([sentence] + pred_scores + true_scores)
```

```python
import pandas as pd
from datasets import load_dataset

traits = ['O', 'C', 'E', 'A', 'N']

dataset = load_dataset("jingjietan/essays-big5", split="train[:20]")
df = dataset.to_pandas()

# Binarize ground truth traits at 0.5 threshold

# Load your predictions
met_df = pd.read_csv("met30.csv")
metsys_df = pd.read_csv("metsys.csv")

# Binarize predictions if needed (if floats)
met_df[traits] = (met_df[traits] >= 0.5).astype(int)
metsys_df[traits] = (metsys_df[traits] >= 0.5).astype(int)

# Optional sanity check: print first few rows to compare
print("Ground truth:\n", df[traits].head())
print("Predictions no system prompt:\n", met_df[traits].head())
print("Predictions with system prompt:\n", metsys_df[traits].head())

def accuracy(pred, true):
    correct = (pred[traits].values == true[traits].values).sum()
    total = true.shape[0] * len(traits)
    return correct / total

print(f"Accuracy no system prompt: {accuracy(met_df, df):.2%}")
print(f"Accuracy with system prompt: {accuracy(metsys_df, df):.2%}")
```
